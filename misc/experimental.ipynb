```python
import subprocess
import sys

# Define list of Panopto lecture URLs
panopto_urls = [
	"https://ntnu.cloud.panopto.eu/Panopto/Pages/Viewer.aspx?id=40ac0657-c340-4ad7-ac37-b264008336e7",
	"https://ntnu.cloud.panopto.eu/Panopto/Pages/Viewer.aspx?id=36050e66-ee41-44eb-af1c-b26400833949",
]

# Output folder where files should be saved
output_folder = "C:\\Users\\Anders\\forelesninger\\styring\\audio\\"

# Format ID for video with audio (update based on `yt-dlp -F` output)
format_id = "9"

# Audio quality
audio_quailty = "64K"


# Loop through URLs and download each lecture

debug_results = []
for url in panopto_urls:
	command = [
		"yt-dlp",
		"-f", format_id,                         # Select correct video+audio format
		"--extract-audio",                       # Extract only audio
		"--audio-format", "mp3",                 # Convert to MP3
		"--audio-quality", audio_quailty,               # Set bitrate
		"--cookies", "cookies.txt",              # Use authentication
		"-o", f"{output_folder}%(title)s.%(ext)s", # Save with title in specified folder
		url
	]
	
	# Run yt-dlp command
	result = subprocess.run(command, capture_output=True, text=True)
	debug_results.append(result)


for result in debug_results:
	print("STDOUT:", result.stdout)
	print("STDERR:", result.stderr)
	print()


```
